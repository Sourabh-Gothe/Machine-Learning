```python
movie_path = "https://raw.githubusercontent.com/abulbasar/data/master/ml-latest-small/movies.csv"
rating_path = "https://raw.githubusercontent.com/abulbasar/data/master/ml-latest-small/ratings.csv"
```

```python
pd.merge
```

```python
movie_df = pd.read_csv(movie_path)
```

```python
rating_df = pd.read_csv(rating_path)
```

```python
movie_df.info()
```

```python
rating_df.info()
```

```python
 Merged_df = pd.merge(movie_df , rating_df , on="movieId")
```

```python
Merged_df.sort_values("rating", ascending=False) [:10]
```

```python
Merged_df.info()
```

```python
rating_agg = rating_df.groupby("movieId").rating.agg([np.mean , len]).reset_index().query("len>=100")
```

```python
Merged_df = pd.merge(movie_df , rating_agg, on="movieId")[["movieId" ,"title" ,"mean"]].sort_values("mean",ascending="False").iloc[:10, :]
```
