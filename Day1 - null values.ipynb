```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"https://raw.githubusercontent.com/abulbasar/data/master/mobile-sales-data.csv")
```

```python
df[df.isnull().sum(axis=1) >0]
```

```python
df.dropna(axis=1)
```

```python
df.fillna(df.mean())
```

```python
df.isnull().sum()
```

```python
df["Salary"] = df.apply(lambda row: row.Age if np.isnan(row.Salary) else row.Salary , axis =1)
```
